ACTIVATION FUNCTION : sigmoid

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))

x = np.linspace(-10,10)
plt.plot(x,sigmoid(x))
plt.title("ACTIVATION FUNCTION : sigmoid")
plt.axis('tight')
plt.show()
```

ACTIVATION FUNCTION : tanh

```python
def tanh(x):
  return np.tanh(x)
x = np.linspace(-10,10)
plt.plot(x,tanh(x))
plt.title('ACTIVATION FUNCTION : tanh')
plt.axis('tight')
plt.show()
```

ACTIVATION FUNCTION : RELU

```python
def max(x):
  li = []
  for i in x:
    if (i > 0):
      li.append(i)
    else:
      li.append(0)
  return li
x = np.linspace(-10,10)
plt.plot(x,max(x))
plt.title("ACTIVATION FUNCTION : RELU")
plt.show()
```
