```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
plt.plot([1,2,4,7,2,1,7])
plt.show()
```

```python
plt.plot([1,2,1,2],[1,2,3,4])
plt.show()
```

```python
x=np.array([1,2,3,1])
y=np.array([2,3,4,5])
plt.scatter(x,y,color=['skyblue','b','r','g'])
plt.show()
```

```python
x=np.array([10,20,30])
y=np.array([10,20,30])
plt.scatter(x,y,color=['g','y','b'])
plt.show()
```

```python
x=np.linspace(2,-2,100)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
x=np.arange(1,6,2)
y=np.arange(2,8,2)
plt.xlabel("number of students")
plt.ylabel("cse")
plt.title('sample')
plt.plot(x,y,c='y',lw=1,marker='*')
plt.show()
```

```python
plt.plot([10,20,10],[20,30,40],ls='dotted',lw=3)
plt.plot([30,50,10],[20,30,40],'r--',lw=2,)
plt.axis([10,30,20,40])
plt.legend(['age','height'])
plt.grid()
plt.show()
```
