```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))

x = np.linspace(-10,10)
plt.plot(x,sigmoid(x))
plt.title("ACTIVATION FUNCTION : sigmoid")
plt.axis('tight')
plt.show()
```
