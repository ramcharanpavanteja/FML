```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
```

```python
data = pd.read_csv("/content/drive/MyDrive/Heart.csv")
```

```python
data
```

```python
data = data.drop(columns = 'Unnamed: 0')
```

```python
data
```

```python
data['ChestPain'] = data['ChestPain'].astype('category')
data['ChestPain'] = data['ChestPain'].cat.codes
data
```

```python
data['Thal'] = data['Thal'].astype('category')
data['Thal'] = data['Thal'].cat.codes
data['AHD'] = data['AHD'].astype('category')
data['AHD'] = data['AHD'].cat.codes
data
```

```python
data.isnull().sum()
```

```python
data=data.dropna()
data
```

```python
X = data.drop(columns = 'AHD')
X
```

```python
y = data['AHD']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=21)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
```

```python
X_test_scaled = scaler.fit_transform(X_test)
```

```python
X_train_scaled
```

```python
X_test_scaled
```

```python
model = LogisticRegression(random_state = 0).fit(X_train_scaled,y_train)
```

```python
model.predict(X_train_scaled)
```

```python
model.score(X_train_scaled,y_train)
```

```python
X_test = scaler.transform(X_test)
y_pred = model.predict(X_test)
```

```python
model.score(X_test, y_test)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
